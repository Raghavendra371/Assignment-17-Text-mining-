## Extracting reviews of shoes from flipkart

```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/birde-combo-pack-2-sports-shoes-running-men/product-reviews/itma4b1034f357f3?pid=SHOGYNMSJZTYFMM9&lid=LSTSHOGYNMSJZTYFMM94HXZWH&marketplace=FLIPKART'
```

```python
page=requests.get(url)
page
```

```python
page.text # Getting text data
```

```python
soup=BeautifulSoup(page.text,'html.parser') # Getting in html format
print(soup.prettify())
```

```python
reviews=soup.find_all('div',class_='_6K-7Co') #picking reviews data 
```

```python
reviews
```

```python
review=[]
for item in reviews:
    item=item.get_text().replace('\n','')
    item=item.strip(' ')
    review.append(item)
```

```python
review
```
